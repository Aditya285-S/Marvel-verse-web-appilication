```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('/content/movies_data.csv')
movies.head()
```

```python
for column in movies.columns:
  if column == 'Movie' or column == 'Movie_id':
    continue
  movies[column] = movies[column].apply(lambda x: x.split() if isinstance(x, str) else x)
```

```python
movies.head(20)
```

```python
movies['tags'] = movies['Description'] + movies['type'] + movies['Tags']
```

```python
new = movies.drop(columns=['Description','type','Tags','Genres','Actors'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
new.head()
```

```python
new['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['Movie'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].Movie)
```

```python
recommend('Deadpool and Wolverine')
```
